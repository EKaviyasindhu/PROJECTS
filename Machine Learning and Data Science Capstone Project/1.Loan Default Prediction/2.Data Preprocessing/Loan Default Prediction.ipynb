```python
##Seperate Numerical and Categorical Columns
def quanQual(dataset):
        quan=[]
        qual=[]
        for columnName in dataset.columns:
            #print(columnName)
            if(dataset[columnName].dtype=='O'): #Alphabet O not zero
                #print("qual")
                qual.append(columnName)
            else:
                #print("quan")
                quan.append(columnName)
        return quan,qual
```

```python
## Check for Outliers
def detect_outliers_iqr(df):
    outlier_summary = {}

    for col in df.select_dtypes(include=['int64','float64']).columns:
        if col == "Default":   # skip target column
            continue
        
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        outliers = df[(df[col] < lower) | (df[col] > upper)]
        outlier_summary[col] = outliers.shape[0]

    return pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Outlier_Count']).sort_values(by="Outlier_Count", ascending=False)

# Run this after loading dataset
outlier_report = detect_outliers_iqr(df)
print(outlier_report)
```

```python
#Result shows that there is no outliers presrnt in this dataset.
```

```python
## Encoding

#Clean column names
df.columns = df.columns.str.strip()        # remove spaces
df.columns = df.columns.str.replace(" ", "")   # remove inner spaces
df.columns = df.columns.str.replace("'", "")   # remove apostrophes
print(df.columns.tolist())

# Binary categorical columns → map Yes/No to 1/0
binary_cols = ['HasMortgage', 'HasDependents', 'HasCoSigner']
for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

# Multi-class categorical columns → One-Hot Encoding
multi_class_cols = ['Education', 'EmploymentType', 'MaritalStatus', 'LoanPurpose']
df = pd.get_dummies(df, columns=multi_class_cols, drop_first=True)

# Convert all boolean dummy columns to int (0/1)
df = df.astype({col: 'int' for col in df.select_dtypes(include=['bool']).columns})
```

```python
df.head()
```

```python
##Scale Numeric Features
#Features like Income, LoanAmount, CreditScore are on different scales.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
num_cols = ['Age','Income','LoanAmount','CreditScore','MonthsEmployed','NumCreditLines','InterestRate','LoanTerm','DTIRatio']
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
##Train-Test Split
from sklearn.model_selection import train_test_split

X = df.drop(columns=["Default"])   # Inputs
y = df["Default"]                  # Output

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```
